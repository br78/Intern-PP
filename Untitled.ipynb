## Inject synthetic fraud (Python)

```python
import pandas as pd

# Step 1: Make sure df_cleaned exists (starting from original df)
df_cleaned = df.copy()

# Step 2: Create a working copy for fraud injection
df_fraud = df_cleaned.copy()

# Step 3: Create upcoding fraud (5% of data)
upcoding = df_fraud.sample(frac=0.05, random_state=42)
df_fraud.loc[upcoding.index, 'diagnosis_code'] = '496'  # Fake COPD code
df_fraud.loc[upcoding.index, 'Tot_Drug_Cst'] *= 2.5     # Inflate drug cost

# Step 4: Create duplicate claims
duplicates = upcoding.copy()
duplicates['claim_id'] = duplicates['PRSCRBR_NPI'].astype(str) + '_DUPE'  # Synthesized claim_id for uniqueness
df_fraud = pd.concat([df_fraud, duplicates], ignore_index=True)

# Step 5: Add 'is_fraud' flag
df_fraud['is_fraud'] = False
df_fraud.loc[upcoding.index, 'is_fraud'] = True
df_fraud.loc[df_fraud['claim_id'].astype(str).str.endswith('_DUPE'), 'is_fraud'] = True

# Step 6: Sanity check
print("Fraud cases count:", df_fraud['is_fraud'].sum())
print("Total rows after fraud injection:", len(df_fraud))
```

# Data Visualization


```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
top_states = df.groupby('Prscrbr_State_Abrvtn')['Tot_Drug_Cst'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(10,6))
sns.barplot(x=top_states.values, y=top_states.index)
plt.title("Top 10 States by Total Drug Cost")
plt.xlabel("Total Drug Cost")
plt.ylabel("State")
plt.show()
```

```python
df['Total_Bene_Gender'] = df['Bene_Feml_Cnt'].fillna(0) + df['Bene_Male_Cnt'].fillna(0)
gender_total = {'Female': df['Bene_Feml_Cnt'].sum(),'Male': df['Bene_Male_Cnt'].sum()}
plt.figure(figsize=(6, 6))
plt.pie(gender_total.values(), labels=gender_total.keys(), autopct='%1.1f%%', colors=['#ff9999','#66b3ff'])
plt.title("Gender Distribution of Beneficiaries")
plt.axis('equal')
plt.show()
```

```python
avg_age_by_state = df.groupby('Prscrbr_State_Abrvtn')['Bene_Avg_Age'].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(10, 6))
sns.barplot(x=avg_age_by_state.index, y=avg_age_by_state.values)
plt.xticks(rotation=90)
plt.title("Average Beneficiary Age by State")
plt.ylabel("Average Age")
plt.xlabel("State")
plt.show()
```

```python
cost_columns = ['Tot_Clms', 'Tot_Drug_Cst', 'Gnrc_Tot_Drug_Cst', 'Brnd_Tot_Drug_Cst', 'Opioid_Tot_Drug_Cst']
corr = df[cost_columns].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Between Drug Cost Metrics")
plt.show()
```

## ML 


```python
features = [
    'Tot_Clms', 'Tot_Drug_Cst', 'Tot_30day_Fills', 'Tot_Day_Suply', 'Tot_Benes',
    'Gnrc_Tot_Drug_Cst', 'Brnd_Tot_Drug_Cst', 'Opioid_Tot_Drug_Cst', 
    'Bene_Avg_Age', 'Bene_Feml_Cnt', 'Bene_Male_Cnt', 'Bene_Avg_Risk_Scre'
]
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# 1. Reduce data size temporarily 
df_sample = df_fraud.sample(n=100000, random_state=42)  

# 2. Preprocessing
features = [
    'Tot_Clms', 'Tot_Drug_Cst', 'Tot_30day_Fills', 'Tot_Day_Suply', 'Tot_Benes',
    'Gnrc_Tot_Drug_Cst', 'Brnd_Tot_Drug_Cst', 'Opioid_Tot_Drug_Cst', 
    'Bene_Avg_Age', 'Bene_Feml_Cnt', 'Bene_Male_Cnt', 'Bene_Avg_Risk_Scre'
]
df_sample[features] = df_sample[features].fillna(0)
X = df_sample[features]
y = df_sample['is_fraud']

# 3. Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.2, random_state=42)

# 4. Logistic Regression
lr = LogisticRegression(max_iter=500, solver='liblinear')  # Faster solver for binary classification
lr.fit(X_train, y_train)
y_pred_lr = lr.predict(X_test)
print("Logistic Regression")
print(classification_report(y_test, y_pred_lr))

# 5. Random Forest (Parallelized & faster)
rf = RandomForestClassifier(n_estimators=50, random_state=42, n_jobs=-1, max_depth=10)
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)
print("Random Forest")
print(classification_report(y_test, y_pred_rf))

# 6. Feature Importance
feat_imp = pd.Series(rf.feature_importances_, index=features).sort_values(ascending=True)
plt.figure(figsize=(10, 6))
feat_imp.plot(kind='barh')
plt.title("Feature Importance (Random Forest)")
plt.tight_layout()
plt.show()
```
